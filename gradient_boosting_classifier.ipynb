```python
from sklearn_helper import liver_dataset

data = liver_dataset()
data.head()
```

### [GradientBoostingClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.GradientBoostingClassifier.html#sklearn.ensemble.GradientBoostingClassifier)

```python
import mlflow
from sklearn.ensemble import GradientBoostingClassifier
from sklearn_helper import split_data, train_model

mlflow.set_experiment( "Gradient Boosting Classifier( liver dataset )" )

X_train, y_train, X_test, y_test = split_data( data, "Diagnosis" )

param_grid = {
    'n_estimators': [ 100, 200, 300 ],
    'learning_rate': [ 0.05, 0.1, 0.2 ],
    'max_depth': [ 2, 3 ],
    'min_samples_split': [ 2, 5, 10 ],
    'min_samples_leaf': [ 1, 2, 4 ],
    'subsample': [ 0.7, 0.8, 0.9, 1.0 ],      # 控制每棵樹的訓練數據比例
    'max_features': [ 'sqrt', 'log2', None ], # 控制每棵決策樹在分裂節點時可以使用多少個最大特徵數量
}

metrics = [ "recall" ]

train_model( GradientBoostingClassifier, param_grid, 
             X_train, y_train, 
             X_test,  y_test, 
             metrics,
             "gradient_boosting", )
```
